```python
# create new df= will contain returns for each bank stock.
returns=pd.DataFrame()
returns
```

```python
# percent_change=(new value - old value) / old value
# computes the change percentage
# we can also called as pct change
import pandas as pd
s=pd.Series([1,2,3,4,5])
s
```

```python
s.pct_change()
```

```python
#we can use pct_change() method on close column to create a column representing this return value.
#create a for loop that goes and for each Bank Stock tickers creates this return column .
tickers=['BAC','C','GS','JPM','MS','WFC']
for tick in tickers:
    returns[tick + ' Return']=bank_stocks[tick]['Close'].pct_change()
```
